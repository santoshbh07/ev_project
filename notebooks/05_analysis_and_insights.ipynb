# Analysis and Insights

This notebook synthesizes findings from exploratory data analysis and
feature-based analysis to examine patterns in electric vehicle adoption
across time, geography, and vehicle type. The goal is to interpret these
patterns and discuss their broader implications.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/ev_population_features.csv")
```

## EV Adoption Is a Recent and Rapid Phenomenon

The EV population is heavily concentrated in recent model years, with
over 80% of vehicles falling into the post-2020 period. Earlier periods
account for a comparatively small share of the dataset, indicating that
widespread EV adoption is a recent development rather than a gradual
long-term trend. This pattern reflects rapid technological maturation,
increased model availability, and expanded policy and market support in
the last decade.


```python
ev_by_year = df.groupby("model_year").size()

plt.figure()
ev_by_year.plot()
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.title("EV Adoption Over Time")
plt.savefig("../figures/ev_adoption_over_time.png", dpi=100, bbox_inches="tight")
plt.show()
```

## Battery Electric Vehicles Dominate the Market

Battery Electric Vehicles account for approximately 80% of the EV
population, substantially outnumbering Plug-in Hybrid Electric Vehicles.
This dominance strengthens in more recent model years, suggesting a
shift away from transitional hybrid technologies toward fully electric
powertrains as battery performance and charging infrastructure improve.


```python
bev_range = (
    df[df["is_bev"]]
    .groupby("model_year")["electric_range"]
    .mean()
)

plt.figure()
bev_range.plot()
plt.xlabel("Model Year")
plt.ylabel("Average Electric Range (miles)")
plt.title("Average Electric Range of BEVs Over Time")
plt.savefig("../figures/avg_electric_range_of_BEVs.png", dpi=100, bbox_inches="tight")
plt.show()
```

## Electric Range Trends Reflect Technological Progress

When analyzed by vehicle type, electric range trends align with expected
technological developments. Battery Electric Vehicles exhibit substantial
range improvements through the late 2010s, while Plug-in Hybrid Electric
Vehicles maintain consistently low electric-only range. Apparent declines
in recent years are attributable to incomplete reporting rather than true
performance regression.


## EV Adoption Is Highly Concentrated Geographically

More than 90% of registered EVs are located in the top ten counties,
demonstrating extreme geographic concentration. This suggests that EV
adoption is strongly associated with urban density, infrastructure
availability, and regional market readiness rather than being evenly
distributed across all regions.


```python
county_counts = df["county"].value_counts().head(10)

plt.figure()
county_counts.plot(kind="bar")
plt.ylabel("Number of EVs")
plt.title("Top 10 Counties by EV Count")
plt.xticks(rotation=45)
plt.savefig("../figures/top_10_counties.png", dpi=100, bbox_inches="tight")
plt.show()
```

## Data Limitations

Several limitations should be noted. Recent model years exhibit
incomplete reporting for certain attributes, including electric range.
Additionally, geographic identifiers reflect registration locations and
may not fully capture vehicle usage patterns. These limitations were
intentionally preserved rather than corrected to avoid introducing
bias or artificial assumptions.


## Synthesis and Implications

Taken together, these findings indicate that EV adoption is a recent,
rapidly accelerating, and highly concentrated phenomenon driven by a
small number of regions and dominated by fully electric vehicles.
Technological improvements in battery range have reinforced the shift
toward BEVs, while geographic concentration highlights the role of
infrastructure and regional readiness. These patterns suggest that
future EV growth will depend not only on continued technological
advancement but also on broader geographic diffusion of charging
infrastructure and policy support.
